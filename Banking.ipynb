```python
bins=[0,100000,300000,float('inf')]
labels = ['Low','Mid','High']

df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
categorical_cols = [
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "Income Band"
]

# Loop through each column and plot distribution
for col in categorical_cols:
    
    print(f"\nValue counts for '{col}':\n")
    print(df[col].value_counts())
    
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
```

```python
categorical_cols = [
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "Income Band"
]

# Loop through columns
for i, predictor in enumerate(categorical_cols):
    
    plt.figure(i, figsize=(7,4))
    
    sns.countplot(data=df, x=predictor, hue='Nationality')
    
    plt.title(f"{predictor} distribution by Gender")
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
```

```python
# Histogram of value counts for different columns except Occupation

for col in categorical_cols:
    
    if col == "Occupation":
        continue
    
    plt.figure(figsize=(8,4))
    
    sns.histplot(data=df, x=col)
    
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    
    plt.show()
```

```python
#numerical analysis
df.columns = df.columns.str.strip()
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

print("Numerical Columns:")
print(numeric_cols)

# Summary statistics
print("\nSummary Statistics:\n")
print(df[numeric_cols].describe())
```

```python
for col in numeric_cols:
    
    plt.figure(figsize=(6,4))
    
    sns.histplot(df[col], bins=20, kde=True)
    
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    
    plt.show()
```

```python
 plt.figure(figsize=(12,8))

sns.heatmap(df[numeric_cols].corr(),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
for col in numeric_cols:
    
    plt.figure(figsize=(6,4))
    
    sns.boxplot(x=df[col])
    
    plt.title(f"Boxplot of {col}")
    plt.show()
   
```

```python
sns.scatterplot(x="Estimated Income", y="Bank Deposits", data=df)

plt.title("Income vs Deposits")
plt.show()
```

```python
pivot_table = pd.crosstab(df['Income Band'], df['Risk Weighting'])

plt.figure(figsize=(8,5))

sns.heatmap(pivot_table,
            annot=True,
            cmap="YlGnBu",
            fmt='d')

plt.title("Income Band vs Risk Weighting")

plt.show()
```

```python
pivot_table = pd.crosstab(df['Occupation'], df['Income Band'])

plt.figure(figsize=(10,6))

sns.heatmap(pivot_table,
            annot=True,
            cmap="viridis",
            fmt='d')

plt.title("Occupation vs Income Band")

plt.show()
```

```python
pivot_table = pd.crosstab(df['GenderId'], df['Amount of Credit Cards'])

plt.figure(figsize=(8,5))

sns.heatmap(pivot_table,
            annot=True,
            cmap="magma",
            fmt='d')

plt.title("Gender vs Number of Credit Cards")

plt.show()
```
